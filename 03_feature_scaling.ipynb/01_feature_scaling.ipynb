### 1. Min Max scaling

```python
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# sample data

data = {'awen_numbers': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
df.head()
```

```python
# scale the data using min max scalar
Scalar = MinMaxScaler()
df['awen_numbers_scaled'] = Scalar.fit_transform(df[['awen_numbers']])
df.head()
```

### 2. standard scalar or Z-score normalization

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# sample data 
data = {'awen_numbers': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
df.head()
```

```python
Scalar = StandardScaler()
df['awen_numbers_scaled'] = Scalar.fit_transform(df[['awen_numbers']])
df.head()
```

### 3. Robust scalar

```python
from sklearn.preprocessing import RobustScaler

# sample data with outliers
data = {'awen_numbers': [10, 20, 30, 1000, 50]}
df = pd.DataFrame(data)

# Robust scaling
Scalar = RobustScaler()
df['awen_numbers_scaled'] = Scalar.fit_transform(df[['awen_numbers']])
print(df)
```

### 4. logrithmic scalling/normalization

```python
import pandas as pd
import numpy as np

# sample data with outliers
data = {'awen_numbers': [100000, 20000, 30000, 1000000, 50000]}
df = pd.DataFrame(data)

# log transform
df['awen_numbers_log'] = np.log(df['awen_numbers'])
df['awen_numbers_log2'] = np.log2(df['awen_numbers'])
df['awen_numbers_log10'] = np.log10(df['awen_numbers'])

df.head()
```

```python
# we make a boxplot of this code to show outlier
import seaborn as sns
#sns.boxplot(df, x= 'awen_numbers')
sns.boxplot(df, x= 'awen_numbers_log10')
```
