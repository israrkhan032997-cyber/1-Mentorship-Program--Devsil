
# LINEAR ALGEBRA IN PYTHON

```python
import numpy as np
```

```python
# coefficient matrix A
A = np.array([[2, 3], [5, 7]])  # 2*2 matrix
A
```

```python
# constant matrix B
B = np.array([8, 20])
B
```

```python
# solve for x
x = np.linalg.solve(A,B)  # solve for x
x
```

# SOLVING COMPLEX EQUATIONS

```python
# COEFFICIENTS MATRIX A WITH COMPLEX NUMBERS
A_complex = np.array([[1+2j, 2-1j],[3-1j, 4]])
A_complex
```

```python
#constants matrix B with complex numbers
B_complex = np.array([3+1j, 3+4j])
B_complex
```

```python
# solving for x in the complex case
X_complex = np.linalg.solve(A_complex, B_complex)
X_complex
```

# (1 + 2i) x + (2 - 1)y = 4 + i ||| (3 - i) x + 4y = 3 + 4i
      
       

```python
# COEFFICIENTS MATRIX A WITH COMPLEX NUMBERS
A_complex = np.array([[1+2j, 2-1j],[3-1j, 4]])

#constants matrix B with complex numbers
B_complex = np.array([3+1j, 3+4j])

# solving for x in the complex case
X_complex = np.linalg.solve(A_complex, B_complex)
X_complex
```

# SINGULAR VALUE DECOMPOSITION IN PYTHON(SVD)


```python
# generate a matrix of 4x3 dataset
A = np.random.rand(4,3)
A
```

```python
# lets apply SVD here
U, sigma, VT = np.linalg.svd(A)
# print each matrix separately
print("U Matrix=", U)
print("---------------------------------")
print("sigma Matrix=", sigma)
print("----------------------------------")
print("VT Matrix=", VT)

```
