```python
# import libraries
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

### Line plots

```python
names = ['israr', 'misbah','noman']
heights = [5.8,5.9,5.0]



# create a plot (line chart)
px.line(x=names, y=heights, title='heights of students')
```

### Bar plots

```python
names = ['israr', 'misbah','noman']
heights = [5.8,5.9,5.0]



# create a plot (line chart)
fig=px.bar(x=names, y=heights, title='heights of students')
fig.show()

# save the plot png file
fig.write_image('./outputs/01_barplot.png')

# save as standalone html
fig.write_html('./outputs/01_barplot.html')
```

### plots on titanic dataset

```python
# import dataset
df = sns.load_dataset('titanic')

# check the head of the dataset
df.head()
```

```python
# scatter plot
fig = px.scatter(df, x='age', y = 'fare',color='sex', size='fare',title= 'plot of titanic dataset')
fig.show()

#save
fig.write_image('./outputs/02_scatterplot.png',scale =3)

# save the plot as a PNG image with 300 DPI resolution
fig.write_image('./outputs/02_scatterplot.png', format='png',width=800, height=600, scale=3)
```

```python
import plotly.graph_objects as go


fig = go.Figure(data=go.Scatter(x=[1, 2, 3, 4], y=[10, 11, 12, 13]))
fig.show()
```

```python
# customize/update the plot 
fig.update_layout(
    title = 'line plot of dummy data',
    xaxis = dict(title='numbers'),
    yaxis = dict(title='values'),
    showlegend = True,
)
```

```python
import numpy as np
import plotly.graph_objects as go

# Generate random data
data = np.random.rand(10, 10)

# Create the heatmap
heatmap = go.Heatmap(z=data)

# Create the figure with the heatmap
fig = go.Figure(data=[heatmap])
fig.show()
```

```python
# heatmap on correlation of titanic dataser of age and fare using plotly

fig = px.imshow(df[['age','fare']].corr())
fig.show()
```
