### SUNBURST PLOT TO SHOW HEIRARCHY

```python
# CREATE SUNBURST PLOT ON TITANIC DATASET
import plotly.express as px
import seaborn as sns
 
# load titanic data
df =sns.load_dataset('titanic')

# create sunburst plot
fig =px.sunburst(df,path=['sex','class','who','alive','alone'],values = 'survived', color='sex')

# show the plot
fig.show()

# save the plot
fig.write_image('./outputs/fig.png',scale=3)
```

```python
import plotly.express as px
import seaborn as sns

# Load tips dataset
tips = sns.load_dataset("tips")

# Create sunburst plot
tip = px.sunburst(
    tips,
    path=["sex","smoker","day", "time", "tip"],
    values="total_bill",
    title="Sunburst Plot of Tips by Day, Time, and Sex"
)

tip.show()
# save the plot
tip.write_image('./outputs/tip.png')
```
