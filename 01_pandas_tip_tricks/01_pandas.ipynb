```python
# import libraries
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data import
df = sns.load_dataset("titanic")
```

```python
#check the data
df.head()
```

MISSING VALUES AND IMPUTING THEM

```python
#FIND OUT THE PERCENTAGE OF MISSING VALUES
df.isnull().sum() / len (df) *100
```

```python
df["age"].mean()
```

```python
df["age"].median()
```

```python
df['age'].mode()[0]
```

```python
#filling missing values of age column by the mean of age
df["age"].fillna(df["age"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
#fill embark and embark_town with mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
#binning of age column into 7 catagories
bins = [0,1,5,12,18,30,50,80]
labels = ['infants','toddlers','kids','teens','youngs','middle aged','old']

#kis column ko bin main convert kerna hai

pd.cut(df['age'], bins=bins, labels=labels)
```

FEATURE ENGINEERING

```python
bins = [0,1,5,12,18,30,50,80]
labels = ['infants','toddlers','kids','teens','youngs','middle aged','old']

#kis column ko bin main convert kerna hai

df['binned_age']=pd.cut(df['age'], bins=bins, labels=labels)
```
