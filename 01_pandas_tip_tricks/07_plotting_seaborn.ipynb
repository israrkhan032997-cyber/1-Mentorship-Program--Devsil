```python
# Data
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# scatter plot
sns.scatterplot(data = df, x = 'age', y = 'fare')
```

```python
# for grouping 
sns.scatterplot(data = df, x = 'age', y = 'fare', hue= 'sex')
```

```python

# for grouping
# scatter plot size of figure size 

plt.figure(figsize=(8,4)) # A4 page size 
sns.scatterplot(data = df, x = 'age', y = 'fare', hue= 'sex')
```

```python
sns.set_theme(style = 'darkgrid',font_scale=1.5)

plt.figure(figsize=(16,8))
sns.scatterplot(data = df, x = 'age', y = 'fare', hue= 'sex')

# change the scatter point colors
palette = {'male': 'blue', 'female': 'red'}
sns.scatterplot(data = df, x = 'age', y = 'fare', hue= 'sex', size='fare',palette=palette)
 
 # labels
plt.xlabel('age')
plt.ylabel('fare')
plt.title('scatter plot of age vs fare')

# saving a plot 300dpi-normally, OR 600dpi-for big screen
plt.savefig('./outputs/scatter_plot.png',dpi=300)

# saving a plot in vector file such as svg
plt.savefig('./outputs/scatter_plot.svg',dpi = 300)







```

### BOX PLOT

```python
df['class'].unique
```

```python
#style of a plot
sns.set_theme(style = 'darkgrid',font_scale=2)

#scatter plot size of figure size
plt.figure(figsize=(14,7)) # A4 page size 

# change the scatter point colors
palette = {'First': 'blue', 'Second': 'red', 'Third': 'yellow'}
sns.boxplot(data = df, x = 'sex', y = 'age', hue= 'class', palette=palette)
 
 # labels
plt.xlabel('sex')
plt.ylabel('age')
plt.title('scatter plot of sex vs age')

#change the legend position and title
#plt.legend(loc='lower left', title='class Name')
plt.legend(loc='upper right', title='class Name')
```

### Boxen plot

```python
#style of a plot
sns.set_theme(style = 'whitegrid',font_scale=2)

#scatter plot size of figure size
plt.figure(figsize=(14,8)) # A4 page size 

# change the scatter point colors
palette = {'First': 'blue', 'Second': 'red', 'Third': 'yellow'}
sns.boxenplot(data = df, x = 'sex', y = 'age', hue= 'class', palette=palette)
 
 # labels
plt.xlabel('Gender')
plt.ylabel('age')
plt.title('Boxenplot plot of Gender vs age')

#change the legend position and title
#plt.legend(loc='lower left', title='class Name')
plt.legend(loc='upper right', title='class Name')

# change the x ticks name
plt.xticks(ticks=[0,1], labels=["Male","Female"], rotation=45)

```

```python
# make scatter plot with plotly
import plotly.express as px

plt.figure(figsize=(12,7))

fig = px.scatter(df, x = 'age', y = 'fare',color='sex')
fig.show()
```

###  BAR PLOT USING SEABORN

```python
#style of a plot
sns.set_theme(style = 'whitegrid',font_scale=2)

#scatter plot size of figure size
plt.figure(figsize=(14,8)) # A4 page size 

# change the scatter point colors
palette = {'First': 'blue', 'Second': 'red', 'Third': 'yellow'}
sns.barplot(data = df, x = 'sex', y = 'age', hue= 'class', palette=palette, capsize=0.1)
 
 # labels
plt.xlabel('Gender')
plt.ylabel('age')
plt.title('Barplot plot of Gender vs age')

#change the legend position and title
#plt.legend(loc='lower left', title='class Name')
plt.legend(loc='upper right', title='class Name')

# change the x ticks name
plt.xticks(ticks=[0,1], labels=["Male","Female"], rotation=45)

```

### violin plot

```python
#style of a plot
sns.set_theme(style = 'whitegrid',font_scale=2)

#scatter plot size of figure size
plt.figure(figsize=(14,8)) # A4 page size 

# change the scatter point colors
palette = {'First': 'blue', 'Second': 'red', 'Third': 'yellow'}
sns.violinplot(data = df, x = 'sex', y = 'age', hue= 'class', palette=palette)
 
 # labels
plt.xlabel('Gender')
plt.ylabel('age')
plt.title('violin plot of Gender vs age')

#change the legend position and title
#plt.legend(loc='lower left', title='class Name')
plt.legend(loc='upper right', title='class Name')

# change the x ticks name
plt.xticks(ticks=[0,1], labels=["Male","Female"], rotation=45)

```
